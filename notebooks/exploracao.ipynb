```python
import os

dataset_path = "../dataset"

print("Resumo do Dataset\n")

for classe in os.listdir(dataset_path):

    pasta = os.path.join(dataset_path, classe)

    if os.path.isdir(pasta):

        arquivos = [
            arq for arq in os.listdir(pasta)
            if arq.lower().endswith(
                (".jpg", ".jpeg", ".png")
            )
        ]

        print(
            f"{classe}: {len(arquivos)} imagens"
        )
```

```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
classe = "flood"

pasta = os.path.join("../dataset", classe)

arquivo = os.listdir(pasta)[0]

img = cv2.imread(os.path.join(pasta, arquivo))

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()
```
